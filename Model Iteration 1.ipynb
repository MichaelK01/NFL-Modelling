<h1> Game By Game Analysis of Model </h1>
This notebook describes the set-up and usage of the Model class. The aim is to perform <br>
ridge regression on an AFL results database to create a model which can consistently predict <br>
outcomes of future games. It is the first iteration in a project to build an algorithm <br>
capable of consistently predicting the score of NFL games. 

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
%matplotlib inline 
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge
```

```python
# Reading in 2019 season data
data = pd.read_csv('spreadspoke_scores.csv', usecols = ['schedule_season', 'team_home', 'score_home', 'score_away', 'team_away'])
data19 = data[data['schedule_season'] == 2019].copy().reset_index(drop = True)
data19.columns = ['Season', 'Home Team', 'Home Score', 'Away Score', 'Away Team']

# Reading in 2019 sack data
sacks_2019 = pd.read_excel('Defense_Stats.xlsx', sheet_name='Passing_Defense_Table_2019', usecols = ['Team', 'Sck'])
sacks_2019.columns = ['Team', 'Sacks']
sacks_2019 = sacks_2019.set_index('Team').sort_index(axis = 0)

# Reading in 2019 1st down rate
downs_2019 = pd.read_excel('Defense_Stats.xlsx', sheet_name='Passing_Offense_Table_2019', usecols = ['Team', '1st'])
downs_2019.columns = ['Team', 'First Down']
downs_2019 = downs_2019.set_index('Team').sort_index(axis = 0)
```

<h3> Model Class Usage </h3>

<h6> Model Attributes: </h6>

1. data: this class attribte needs to be a pandas dataframe containing the following headings: 'Home Score', 'Away Score', 'Home Team', 'Away Team'. 
2. Model_Type: (optional) This sets the method used to calculate the ratings. 1, 2, 3. Default of 1
3. Sack_Data: (required for Model_Type 2 or 3) must contain full team names as well as 'Sacks' column
4. First_Down: (required for Model_Type 2 or 3) must contain full team names as well as 'First Down' column
5. alpha: (optional) set alpha value for ridge regression. Default of 6
        

```python
class Model:
    
    def __init__(self, data, Model_Type = 1, Sack_Data = None, First_Down = None, alpha = 6):
        
        self.data = data
        self.Sack_Data = Sack_Data
        self.First_Down = First_Down
        self.Model_Type = Model_Type
        self.alpha = alpha
        self.df_mod = None
        
        if self.Model_Type == 1:
            self.ratings = self.Ratings1()
        elif self.Model_Type == 2:
            self.ratings = self.Ratings2(Sack_Data, First_Down)
        elif self.Model_Type == 3:
            self.ratings = self.Ratings3(Sack_Data, First_Down)
        else:
            print('Invalid Model Type')
           
    def Ratings1(self):

        self.data['Goal Diff'] = self.data['Home Score'] - self.data['Away Score']        
        df_vis = pd.get_dummies(self.data['Away Team'], dtype=np.int64)
        df_home = pd.get_dummies(self.data['Home Team'], dtype=np.int64)
        df_mod = df_home.sub(df_vis) 
        df_mod['Goal Diff'] = self.data['Goal Diff']
        
        self.df_mod = df_mod
        
        lr = Ridge(alpha = self.alpha) 
        X = df_mod.drop(['Goal Diff'], axis=1)
        y = df_mod['Goal Diff']
        lr.fit(X, y)

        ratings = pd.DataFrame(data={'Team': X.columns, 'Rating': lr.coef_})
        ratings = ratings.sort_values('Rating', ascending = False).reset_index(drop = True).set_index('Team')
        
        return ratings
    
    def Ratings2(self, Sack_Data, First_Down):

        self.data['Goal Diff'] = self.data['Home Score'] - self.data['Away Score']        
        df_vis = pd.get_dummies(self.data['Away Team'], dtype=np.int64)
        df_home = pd.get_dummies(self.data['Home Team'], dtype=np.int64)
        df_mod = df_home.sub(df_vis) 
        
        df_sack = df_mod.copy() 
        for i in Sack_Data.index:
            df_sack.loc[:,i] *= (1/(Sack_Data.loc[i, 'Sacks'] / Sack_Data['Sacks'].mean()))
         #   df_sack.loc[:,i] *= (Sack_Data.loc[i, 'Sacks'] - Sack_Data['Sacks'].mean())
        
        df_mod = pd.concat([df_mod, df_sack], axis=1)
        
        df_down = df_mod.copy() 
        for i in First_Down.index:
            df_down.loc[:,i] *= (1/(First_Down.loc[i, 'First Down'] / First_Down['First Down'].mean()))
         #   df_down.loc[:,i] *= (First_Down.loc[i, 'First Down'] - First_Down['First Down'].mean())
        df_mod = pd.concat([df_mod, df_down], axis=1)
        df_mod['Goal Diff'] = self.data['Goal Diff']
        
        self.df_mod = df_mod
        
        lr = Ridge(alpha= self.alpha) 
        X = df_mod.drop(['Goal Diff'], axis=1)
        y = df_mod['Goal Diff']
        lr.fit(X, y)

        rat = pd.DataFrame(data={'Team': X.columns, 'Rating': lr.coef_})
        for i in range(32):
            rat.loc[i,'Rating'] = rat.loc[i, 'Rating'] + rat.loc[i + 32, 'Rating']/(1/(Sack_Data.loc[rat.loc[i, 'Team'], 'Sacks'] / Sack_Data['Sacks'].mean())) + rat.loc[i + 64, 'Rating']/(1/(First_Down.loc[rat.loc[i,'Team'], 'First Down'] / First_Down['First Down'].mean()))
            #rat.loc[i,'Rating'] = rat.loc[i, 'Rating'] + (rat.loc[i + 32, 'Rating']/(Sack_Data.loc[rat.loc[i,'Team'], 'Sacks'] - Sack_Data['Sacks'].mean())) + (rat.loc[i + 64, 'Rating']/(First_Down.loc[rat.loc[i,'Team'], 'First Down'] - First_Down['First Down'].mean()))
        ratings = rat.loc[0:31,:].copy()
        ratings = ratings.sort_values('Rating', ascending = False).reset_index(drop = True).set_index('Team')
        
        return ratings
    
    def Ratings3(self, Sack_Data, First_Down):

        self.data['Goal Diff'] = self.data['Home Score'] - self.data['Away Score']        
        df_vis = pd.get_dummies(self.data['Away Team'], dtype=np.int64)
        df_home = pd.get_dummies(self.data['Home Team'], dtype=np.int64)
        df_mod = df_home.sub(df_vis) 
        
        df_mod['First Down'] = np.zeros(df_mod.shape[0])
        df_mod['Sacks'] = np.zeros(df_mod.shape[0])
        for i in range(df_mod.shape[0]):
            h = 0
            a = 0
            for j in Sack_Data.index:
                if df_mod.loc[i,j] == 1:
                    h = j
                elif df_mod.loc[i,j] == -1:
                    a = j
            df_mod.loc[i, 'Sacks'] = df_mod.loc[i, 'Sacks'] + (Sack_Data.loc[h,'Sacks'] - Sack_Data.loc[a,'Sacks'])
            df_mod.loc[i, 'First Down'] = df_mod.loc[i, 'First Down'] + (First_Down.loc[h,'First Down'] - First_Down.loc[a,'First Down'])
        
        df_mod['Goal Diff'] = self.data['Goal Diff']
        df_mod.loc[:, 'Sacks'] /= df_mod.loc[:, 'Sacks'].std()
        df_mod.loc[:, 'First Down'] /= df_mod.loc[:, 'First Down'].std()
        
        self.df_mod = df_mod
        
        lr = Ridge(alpha = self.alpha) 
        X = df_mod.drop(['Goal Diff'], axis=1)
        y = df_mod['Goal Diff']
        lr.fit(X, y)

        ratings = pd.DataFrame(data={'Team': X.columns, 'Rating': lr.coef_}) 
        ratings = ratings.sort_values('Rating', ascending = False).reset_index(drop = True).set_index('Team')
        
        return ratings
    
    def Winner(self, Home, Away):
        
        if self.Model_Type == 3:
            pred_points_diff = self.ratings.loc[Home, 'Rating'] - self.ratings.loc[Away, 'Rating'] + self.ratings.loc['Sacks', 'Rating']*(self.Sack_Data.loc[Home,'Sacks'] - self.Sack_Data.loc[Away,'Sacks'])/(self.df_mod.loc[:, 'Sacks'].std()) + self.ratings.loc['First Down', 'Rating']*(self.First_Down.loc[Home,'First Down'] - self.First_Down.loc[Away,'First Down'])/(self.df_mod.loc[:, 'First Down'].std())

            if pred_points_diff > 0:
                return [pred_points_diff, Home]
            else:
                return [pred_points_diff, Away]
            
        else:
            pred_points_diff = self.ratings.loc[Home, 'Rating'] - self.ratings.loc[Away, 'Rating']
            if pred_points_diff > 0:
                return [pred_points_diff, Home]
            else:
                return [pred_points_diff, Away]
    
    def Accuracy(self, start = 100, display_data = False, plot = False):
        
        df = self.data.copy()
        df['Winner'] = np.where(df['Goal Diff'] > 0, df['Home Team'], df['Away Team'])
        
        tot_suc = 0
        tot_fail = 0
        games = df.shape[0]
        games_predicted = games - start
        run_suc_rate = {}
        run_suc = {}
        pred_points_dev = {}
        
        for i in range(start, games-1):
            mod = Model(df.loc[:i,:].copy(), self.Model_Type, self.Sack_Data, self.First_Down, self.alpha)
            pred_points_dev[i+1] = np.abs(mod.Winner(df.loc[i+1, 'Home Team'], df.loc[i+1, 'Away Team'])[0] - self.data.loc[i+1,'Goal Diff'])
            if df.loc[i+1, 'Winner'] == mod.Winner(df.loc[i+1, 'Home Team'], df.loc[i+1, 'Away Team'])[1]:
                tot_suc += 1
                run_suc[i+1] = tot_suc
                run_suc_rate[i+1] = (tot_suc*100)/(tot_suc + tot_fail)
            else:
                tot_fail += 1
                run_suc[i+1] = tot_suc
                run_suc_rate[i+1] = (tot_suc*100)/(tot_suc + tot_fail)
                
        fin_suc = (tot_suc*100)/(tot_suc + tot_fail)
        sum_suc = 0
        for i in run_suc_rate.values():
            sum_suc += i
        
        avg_suc = sum_suc/(games - start)
        mean_points_dev = np.mean(list(pred_points_dev.values()))
        median_points_dev = np.median(list(pred_points_dev.values()))
        q75, q25 = np.percentile(list(pred_points_dev.values()), [75 ,25])
        iqr_points_dev = q75 - q25
        
        if plot:
            
            fig, (g1, g2, g3) = plt.subplots(1, 3, figsize=(12,4), facecolor='black')

            g1.plot(list(run_suc_rate.keys()) , list(run_suc_rate.values()))
            g1.grid()
            g1.set_title('Scatter Plot')
            g1.set_xlabel('Game Number')
            g1.set_ylabel('Prediction Success Rate')

            g2.plot(list(run_suc.keys()) , list(run_suc.values()))
            g1.grid()
            g2.set_title('Scatter Plot')
            g2.set_xlabel('Game Number')
            g2.set_ylabel('Total Successes')
            
            g3.scatter(x = list(pred_points_dev.keys()), y = list(pred_points_dev.values()))
            g1.grid()
            g3.set_title('Scatter Plot')
            g3.set_xlabel('Game Number')
            g3.set_ylabel('Prediction Success Rate')
            
            fig.tight_layout()
            plt.show()
            
        if display_data:
            print('->  No. of games predicted: {}'.format(games_predicted),
                  '\n->  No. of successes: {}'.format(tot_suc),
                  '\n->  No. of failures: {}'.format(tot_fail),
                  '\n->  Average running success: {}'.format(avg_suc),
                  '\n->  Final success rate: {}'.format(fin_suc),
                  '\n->  Mean points model was off by: {}'.format(mean_points_dev),
                  '\n->  Median points model was off by: {}'.format(median_points_dev),
                  '\n->  IQR points model was off by: {}'.format(iqr_points_dev) )

        return (games_predicted, tot_suc, tot_fail, fin_suc, avg_suc, run_suc_rate, run_suc, pred_points_dev, mean_points_dev, median_points_dev, iqr_points_dev)
    
```

```python
## EXAMPLE USAGE ##

# creating model instance
mod_2019_type1 = Model(data19, 1)
# display ratings
display(mod_2019_type1.ratings.head(), mod_2019_type1.ratings.tail())
# make prediction
print('Predicted winner between 49ers and Ravens: {}'.format(mod_2019_type1.Winner('San Francisco 49ers', 'Baltimore Ravens')))
print('Predicted winner between Ravens and 49ers: {}'.format(mod_2019_type1.Winner('Baltimore Ravens', 'San Francisco 49ers')))
# display model accuracy, along with stats and plots (note the arguments are optional)
acc = mod_2019_type1.Accuracy(start = 100, display_data = True, plot = True)
```

```python
# EXAMINING HOW CHANGING ALPHA EFFECTS THE RESULTS

alpha = np.linspace(1500, 3000, 15)
dev = {}
rat = {}

for i in alpha:
    mod = Model(data19, 3, sacks_2019, downs_2019, alpha = i)
    acc = mod.Accuracy(start = 100, display_data = False, plot = False)
    rat[i] = acc[3]
    dev[i] = acc[8]

fig, (g1, g2) = plt.subplots(1, 2, figsize=(12,4), facecolor='black')

g1.plot(list(dev.keys()) , list(dev.values()))
g1.grid()
g1.set_title('Mean Deviation')
g1.set_xlabel('Alpha')
g1.set_ylabel('Deviation')

g2.plot(list(rat.keys()) , list(rat.values()))
g1.grid()
g2.set_title('Success Rate')
g2.set_xlabel('Alpha')
g2.set_ylabel('Rate')

plt.tight_layout()
plt.show()
```
