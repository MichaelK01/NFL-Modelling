### Section 2 - Statement of Candidate Models

Seven simple linear regression models are explored to predict score difference in a given game. Data is restricted to the scores of previous games in the given season. The NFL is an interesting setting for such a regression problem since the effect of the 'Draft' means that teams which are weak one season can be very strong the next. This makes historical data unreliable and forces us to concentrate on the season of interest. As such, data is limited, and dimensionality becomes and issue.

- <b> Model 1 a </b>: $ y = \beta_0X_0 + \beta_1X_1 + ... + \beta_{31}X_{31} $ <br>
Where $X_i$ is a dummy variable with value 1 if team $i$ is the home team, -1 if away, or 0 otherwise. Dropping the intercept removes perfect multicollinearity, and also makes sense since if no team were playing there would be no score. Model coefficients interpreted as the effect on score difference a team will have in a game if they are playing. Ordinary Least Squares (OLS) is performed here to give an unbiased estimation of the $\beta_i$<br>
- <b>Model 1 b</b>: $ y = \beta_0X_0 + \beta_1X_1 + ... + \beta_{31}X_{31} $ <br>
Ridge Regression is performed here, instead of OLS. The loss function in this case is given by: $ ||y - X\beta|| + \alpha||\beta||^2 $, creating a biased estimator for $\beta$. Alpha is determined using an 5-fold validated gridsearch method, where the MSE statistic is being optimised. Althought the purpose of Ridge Regression is traditionally to reduce variance of the coefficients due to multicollinearity from the predictor variables, it is used in this case to penalize large coefficients. This regularization is intended to penalize overfitting due to the very small number of observations available to train the model. Irrespective of dimensionality, introducing a bias has been shown by Hoerl and Kennard (1970) to always reduce the MSE for some $\alpha > 0$. Therefore if bias isn't an issue and prediction power is the only model use, ridge regression can't hurt.
- <b>Model 2 a,b</b>: $ y = \beta_0 + \beta_1X_1 + ... + \beta_{31}X_{31} $ <br>
In this model a reference category is dropped (Atlantic Falcons) and the intercept term is included. Note the reference category is dropped since otherwise, inclusion of the intercept term would create perfect multicollinearity. This removes the symmetry problem of the first two models, that is a home advantage can be included in $\beta_0$. Again, $X_i$ is a dummy variable with value 1 if team $i$ is the home team, -1 if away, or 0 otherwise. Including the intercept complicates the interpretation of the model coefficients somewhat, and it is not clear if this model is useful. Coefficients for Model 2 a are calculated using OLS, while 2 b uses ridge regression.<br>
- <b>Model 3 a,b</b>: $ y = \beta_0X_0 + \beta_1X_1 + ... + \beta_{63}X_{63} $ <br>
Where $X_i$ is a dummy variable with value 1 if team $i$ is home for $1\leq i\leq31$, away for $32\leq i\leq63$, or 0 otherwise. Coefficients are again calculated using Ridge regression, to penalise for overfitting. This model completely removes the symmetry problem by giving each team a seperate coefficient for home games and away games. Coefficients for Model 3 a are calculated using OLS, while 3 b uses ridge regression. Dimensionality may pose a problem in this model, and it is expected that the ridge regression will outperform OLS.<br>
- <b>Model 4</b>: $ y =  \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_{32}X_{32} $ <br>
This model includes all dummy variables and the intercept term. Perfect Multicollinearity exists here so OLS Cannot be used since $X$ is not full-rank. Therefore ridge regression is used to bias the data. It will be interesting to see if a large bias is needed to generate a stable set of coefficients. If ridge regression is effective then this may be the optimum model, as it incorporates a 'home advantage' coefficient ($\beta_0$) as well as a single 'rating' (coefficient) for each team. This model has half the dimensions of Model 3, and the added bonus of asymmetry over Model 1.


### Section 1 - Exploratory Data Analysis & Data Preparation

```python
# Reading in 2019 season data
data = pd.read_csv('spreadspoke_scores.csv', usecols = ['schedule_season', 'team_home', 'score_home', 'score_away', 'team_away'])
data19 = data[data['schedule_season'] == 2019].copy().reset_index(drop = True).drop('schedule_season', axis=1)
data19.columns = ['Home Team', 'Home Score', 'Away Score', 'Away Team']
data19['Goal Diff'] = data19['Home Score'] - data19['Away Score']
```

```python
# Preprocessing Data for models listed in Section 2
df_vis = pd.get_dummies(data19['Away Team'], dtype=np.int64)
df_home = pd.get_dummies(data19['Home Team'], dtype=np.int64)

# Model 1, a and b & Model 4
df1 = df_home.sub(df_vis) 
df1['Goal Diff'] = data19['Goal Diff']
X1 = df1.drop('Goal Diff', axis=1)

# Model 2, a and b
df2 = df_home.sub(df_vis) 
df2['Goal Diff'] = data19['Goal Diff']
df2.drop('Arizona Cardinals', axis = 1, inplace = True)
X2 = df2.drop('Goal Diff', axis=1)

# Model 3, a and b
df3 = pd.concat([df_home.add_suffix('_home'), df_vis.add_suffix('_vis')], axis=1)
df3['Goal Diff'] = data19['Goal Diff']
X3 = df3.drop('Goal Diff', axis=1)

y = data19['Goal Diff']
```

```python
# Exploratory Data Analysis of Predictor Variables

display(data19.head())
print('No. of Games in 2019 Season: ', data19.shape[0])
print('No. of teams in contention: ', len(np.unique(data19['Home Team'])))

# create dictionary object with stats for each team
teams = {}
for i in np.unique(data19['Home Team']):
    # stats in array: wins, losses, number of home games played, and number of away games played
    teams[i] = [0,0,0,0]

for team in teams:
    for i in range(data19.shape[0]):
        if data19['Home Team'][i] == team:
            teams[team][2] += 1
            if data19['Goal Diff'][i] > 0:
                teams[team][0] += 1
            elif data19['Goal Diff'][i] < 0:
                teams[team][1] -= 1
        elif data19['Away Team'][i] == team:
            teams[team][3] -= 1
            if data19['Goal Diff'][i] < 0:
                teams[team][0] += 1
            elif data19['Goal Diff'][i] > 0:
                teams[team][1] -= 1

# plotting Results
results_tab = pd.DataFrame(teams.values(), teams.keys()).sort_values(0, ascending = False)

plt.figure(figsize=(10, 7))
plt.bar(results_tab.index.values, results_tab[0], color='#06d6a0', label='Wins')
plt.bar(results_tab.index.values, results_tab[1], color='#e63946', label='Losses')
plt.title('Number of Wins & Losses 2019')
plt.xticks(rotation = 90)
plt.legend()
plt.show()

plt.figure(figsize=(10, 7))
plt.bar(results_tab.index.values, results_tab[2], color='#ffafcc', label='Home')
plt.bar(results_tab.index.values, results_tab[3], color='#a2d2ff', label='Away')
plt.title('Number of Home & Away Games 2019')
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
# Correlation Matrix for Predictor Variables

f = plt.figure(figsize=(14, 12))
plt.matshow(X1.corr(), fignum=f.number)
plt.yticks(range(X1.select_dtypes(['number']).shape[1]), X1.select_dtypes(['number']).columns, fontsize=12)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=12)
plt.title('Correlation Matrix of Predictor Variables', fontsize=18);
```

```python
# Variance Inflation Factors for predictor variables

# There are two different sets of variables: Models 1,2,4 contain 31 dummies, while Model 3 contains 63

# Reference catagory dropped for each set otherwise the predictor variables would be perfectly collinear
Xcor1 = X1.drop(['Arizona Cardinals'], axis=1)
Xcor2 = X3.drop(['Arizona Cardinals_home'], axis=1)

# Models 1,2,4
vif_data1 = pd.DataFrame()
vif_data1["Predictor"] = Xcor1.columns
vif_data1["VIF"] = [variance_inflation_factor(Xcor1.values, i) for i in range(len(Xcor1.columns))]
display(vif_data1)

# Model 3
vif_data2 = pd.DataFrame()
vif_data2["Predictor"] = Xcor2.columns
vif_data2["VIF"] = [variance_inflation_factor(Xcor2.values, i) for i in range(len(Xcor2.columns))]
display(vif_data2)
```

The correlation matrix indicates no significant linear relationship between any two independent variables.
The maximum variance inflation factors are 2.486 and 2.675, in the first and second set of predictor variables respectively. These VIFs are not far from the minimum of 1, indicating that multicollinearity between the predictor variables is not of significant concern in the regression model.

### Section 4 - Constructing Models

In this section, the models are instanciated and trained on a training data set. A prediction is then made on the testing set, and the Mean Squared Error (MSE) and Root Mean Squared Error (RMSE, the mean absolute error) are calculated. The $R^2$ statistic is also calculated, although this is not of primary concern. Finally, the model is trained on the full dataset and model coefficients are extracted.

```python
# Custom grid searcher which iterates GridSearchCV from sklearn to give a precise estimate of the optimum alpha
# runs 'l' loops over an inital grid from 'a' to 'b' at intervals of 's'

def LoopedGridSearchCV(mod, X, y, a=0.1, b=11, s=0.5, l=3):
    alpha = 0
    param_grid = {'alpha' : np.arange(a, b, s)}
    yhat = (y - np.mean(y))/np.sqrt(np.var(y))
    for i in range(1, l):
        alpha = GridSearchCV(mod, param_grid, scoring='neg_mean_squared_error', cv=5).fit(X, yhat).best_params_['alpha']
        param_grid = {'alpha' : np.arange(alpha - 2**(-i), alpha + 2**(-l), 10**(-i))}
    return alpha
```

```python
# Creating one train test split to use for models with same predictor set for consistency
# Many articles indicate that a 70/30 train test ratio is ideal for smaller datasets
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y, test_size=0.3)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y, test_size=0.3)
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y, test_size=0.3)
```

```python
LinMod1a_test = LinearRegression(fit_intercept = False).fit(X_train1, y_train1)
print('Test Set Stats:\n- R Squared: ', LinMod1a_test.score(X_test1, y_test1))

# calculating RSS on test set
y_pred1a = LinMod1a_test.predict(X_test1)
MSE1a = np.mean((y_test1 - y_pred1a)**2)
print('- MSE: ', MSE1a)
print('- RMSE: ', MSE1a**(1/2) )
# Training final model of full dataset
LinMod1a = LinearRegression(fit_intercept = False).fit(X1, y)

# Test to make sure reasonable results are being produced
print('\nMaking a single prediction: ', LinMod1a.predict(X1.loc[100,:].values.reshape(1,-1)))
print('Actual goal diff that game: ', y[100])

# extracting and displaying parameters
ratings = pd.DataFrame(data={'Team': X1.columns, 'Rating': LinMod1a.coef_})
ratings = ratings.sort_values('Rating', ascending = False).reset_index(drop = True).set_index('Team')
display(ratings.head())
```

#### Identification and Treatment of Outliers

The models above all contain observations with unusually large observations. The exclusion of these observations from the model training could produce more accurate results. Exclusion of outliers in this case is acceptable since the model is used for predictive purposes only, and real world outliers (upsets in games) are relatively common. A simple test for outliers is conducted using standardized residuals:
$$ z_i = \frac{y_i - \hat{y_i}}{\sqrt{s}} $$
Where $s$ is the MSE. If the standardized residual is greater than or equal to 3 it is excluded. Although somewhat arbitrary, this is a standard practice (Yan, X., Su, X. 2009). It was shown in the previous section that the residuals are approximately normally distributed, therefore 99.7% of the data should be within 3 standard deviations. To test if this technique is useful the model is re-run on the same training set after outliers are removed, then MSE and quantiles are calculated again on the test set to see if there is a drop. 

##### Model 1 a

```python
r1a = y_train1 - LinMod1a_test.predict(X_train1)
z1a = (r1a - np.mean(r1a))/(np.sum((r1a - np.mean(r1a))**2)/len(y_train1))**(1/2)
print('Standardized Residual Variance: {}, and mean {}'.format(np.var(z1a), np.mean(z1a)))
z1a = z1a.to_frame()
print('Origional No. of observations in training set: ', z1a.shape[0])

z1a.drop(z1a[np.abs(z1a['Goal Diff']) > 3].index, inplace = True)
print('Final No. of observations in training set: ', z1a.shape[0])
newTrainingIndexes1a = z1a.index.tolist()

LinModnew1a = LinearRegression(fit_intercept = False)
LinModnew1a.fit(X_train1.loc[newTrainingIndexes1a,:], y_train1[newTrainingIndexes1a])
print('Test Set Stats:\n- R Squared: ', LinModnew1a.score(X_test1, y_test1))

# calculating new MSE on test set
y_prednew1a = LinModnew1a.predict(X_test1)
MSEnew1a = np.mean((y_test1 - y_prednew1a)**2)
print('- Old MSE: ', MSE1a)
print('- New MSE: ', MSEnew1a )
```

It is noted that no observations are removed. This is likely due to the low number of observations in the set. Another method may be to adjust the lower threshold of outliers away from 3, or a t-distribution could be assumed which is often more useful for smaller datasets. These will not be explored in this notebook.

### Section 6 - Game by Game Analysis of Model Performance

The primary statistic of interest is mean squared error. In this section the model is run for the first n games in the 2019 season, and the n+1th game is then predicted. This is process begins at n = 34, and runs for the entire season. This gives a two week sit-out period where no predictions are made. Each team plays twice during this period. Further investigation of the sit-out period is warrented as n = 34 is arbitrary.

Note: the model will be run many times, making it unfeasable to calculate a precise ridge hyperparameter each time. A rough looped GridSearchCV function (Defined previously as LoopedGridSearchCV) is deployed to estimate alpha.

Finally, for Model 1a, a comparison between an algorithm with outlier removal and without outlier removal is made.  

##### Model 1 a 

```python
outliers1a = {}
results1a_out = {}
results1a = {}

for i in range(34, df1.shape[0]-1):
    
    LinModIterTest1a = LinearRegression(fit_intercept = False).fit(X1.loc[:i,:], y.loc[:i])
    # adding prediction to no outlier analysis
    results1a[i] = LinModIterTest1a.predict(X1.loc[i+1,:].to_numpy().reshape(1,-1))
    # calculating prediction for outlier analysis
    r1a = y.loc[:i] - LinModIterTest1a.predict(X1.loc[:i,:])
    z1a = ((r1a)/((np.sum(r1a**2)/(len(y.loc[:i])-1))**(1/2))).to_frame()
    z1a.drop(z1a[np.abs(z1a['Goal Diff']) > 3].index, inplace = True)
    newTrainingIndexes1a = z1a.index.tolist()
    LinModIter1a = LinearRegression(fit_intercept = False)
    LinModIter1a.fit(X1.loc[newTrainingIndexes1a,:], y[newTrainingIndexes1a])
    # adding prediction for outlier analysis
    outliers1a[i] = len(y.loc[:i]) - len(newTrainingIndexes1a)
    results1a_out[i] = LinModIter1a.predict(X1.loc[i+1,:].to_numpy().reshape(1,-1))

indexes = np.array(list(results1a.keys()))
actual = y[35:].to_numpy().reshape(-1,1)    

prediction1a_out = np.array(list(results1a_out.values()))
abs_diff1a_out = np.abs(prediction1a_out - actual)
prediction1a = np.array(list(results1a.values()))
abs_diff1a = np.abs(prediction1a - actual)

# Graphing results
print('Note: In the figure below \'OR\' stands for Outliers Removed')
plt.figure(figsize=(14, 7))
plt.scatter(indexes, abs_diff1a_out, color='#06d6a0', label='Outliers Removed')
plt.scatter(indexes, abs_diff1a, color='#e63946', label='Outliers Remain')

plt.axhline(np.quantile(abs_diff1a_out, .95), linestyle='--', 
            color='b', label='95th quantile OR: {}'.format(round(np.quantile(abs_diff1a_out, .95),2)))
plt.axhline(np.quantile(abs_diff1a_out, .50), linestyle='--',
            color='g', label='50th quantile OR: {}'.format(round(np.quantile(abs_diff1a_out, .5),2)))
plt.axhline(np.quantile(abs_diff1a, .95), linestyle='-',
            color='b', label='95th quantile: {}'.format(round(np.quantile(abs_diff1a, .95),2)))
plt.axhline(np.quantile(abs_diff1a, .50), linestyle='-',
            color='g', label='50th quantile: {}'.format(round(np.quantile(abs_diff1a, .5),2)))

## toggle two comments below for rolling average
#plt.plot(indexes, pd.Series(abs_diff1a_out.flatten()).rolling(7).mean(), 'c-', label='3 week moving')
#plt.plot(indexes, pd.Series(abs_diff1a.flatten()).rolling(7).mean(), 'c--', label='3 week moving avg OR')

plt.title('Prediction Error - Model 1 a')
plt.ylabel('Points')
plt.xlabel('Game')
plt.legend(title = 'Legend', bbox_to_anchor = (1.05, 1), loc = 'upper left')
plt.show()

# Toggle comment to view number of outliers
#print(outliers1a)
```


Note: if the maximum alowed deviation (3) is adjusted to 2.85 then this model beats the model which doesn't remove outliers. Although this adjustment was made with hindsight to minimise prediction error (i.e. it could not have been done on a live trial) there is potential to examine this value over a number of years of historical results to come up with an estimated best parameter.

The model only removed outliers in the first half of the data, and in each case only one observation was removed. This may not be ideal, since it is not clear what is relevant or not on a poorly trained model. Also, there are a number of observations around the 150 mark which appear to be quite large. It is not immediately clear from this table which model may be deemed best. However, It is noted that all of the regularized models outperformed their ordinary least squares counterparts. The OLS models can thus be removed from consideration.

Model 4 performed significantly better than the other models, especially with respect to the 95th quantile. One concern with this model is the variance of its coefficients. Before applying the ridge bias, the model is perfectly collinear, and its not clear to what extent the bias fixes this. Further investigation needs to be conducted into this, with computation of eigenvalues of $X'X$, which are proportional to the variance of the estimated coefficients (Hoerl, A. E., Kennard, R. W. 1970, and Marquardt, D. W., Snee, R. D. 1975). 

Models 1 b and 2 b perform very similarly, with a marginally lower 50th quantile for 2 b but a larger 95th quantile for 2 b. Of these two models, model 1 b has a much simpler interpretation and gives each team one 'rating' irrespective of playing home or away. The reference category for model 2 b, Arizona Cardinals, may be giving poor predictions when they are playing. For these reasons, model 1 b is deemed more useful than model 2 b.

Finally, model 3 b performed worst of the four regularized models. If there were more data points, I would expect this model to perform better but dimensionality proved too much of an issus for the small dataset. The tradoff of a unique home and away score for each team for doubling the dimensionality of the model appears to be too expensive. 

In conclusion, model 4 and model 1 b have to most potential for application. More research needs to be done before it is identified wether the added intercept term is worth the added variance due to multicollinearity. 

It is noted that these models are simple. They only use game results in a given season to predict future results. Two obvious extensions are worth exploring:
- The models may benefit from the inclusion of more regressors such as team performance stats, injuries etc. Incorporating these into the model will require an analysis of multicolinearity using Generalised Variance Inflation Factors (Fox, Monette 1992) due to the qualitative nature of the current regressors.
- Interpretting the models in a Bayesian setting, and including priors about team coefficients could be very benificial. Teams which have historically been weaker could have stronger regularization than stronger teams.
